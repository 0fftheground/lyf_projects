```python
#高德地图地理编码开始
import requests
def geocoding(address):
    """
    http://api.map.baidu.com/geocoding/v3/?address=北京市海淀区上地十街10号&output=json&ak=您的ak&callback=showLocation //GET请求
    """
    base_url = "https://restapi.amap.com/v3/geocode/geo?address={}&key=c33b02b68bd7785d75ed05f0178a8e05".format(address)
    response = requests.get(base_url)
    answer = response.json()
    if answer['status'] == '1' and len(answer['geocodes'])>0:     
        longit,latit = answer['geocodes'][0]['location'].split(',')
    else:
        print(address)
        print(answer)
        longit = 0
        latit = 0
    return latit, longit
```
